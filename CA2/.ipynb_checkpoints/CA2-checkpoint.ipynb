```python
from PIL import Image, ImageTk
import tkinter as tk 
from tkinter import messagebox 

food_prices = {
    "JOLLOF RICE": 350,
    "COCONUT FRIED RICE": 350,
    "JOLLOF SPAGHETTI": 350,
    "SWEET CHILLI CHICKEN": 1100,
    "GRILLED CHICKEN WINGS": 400,
    "FRIED BEEF": 400,
    "FRIED FISH": 500,
    "BOILED EGG": 200,
    "SAUTEED SAUSAGES": 200,
    "SAVOURY BEANS": 350,
    "ROASTED SWEET POTATOES": 300,
    "FRIED PLANTAINS": 150,
    "MIXED VEGETABLES SALAD": 150,
    "BOILED YAM": 150,
    "EBA": 100,
    "POUNDO YAM": 100,
    "SEMO": 100,
    "ATAMA SOUP": 450,
    "EGUSI SOUP": 480,
    "WATER": 200,
    "GLASS DRINK(35CL)": 150,
    "PET DRINK(35CL)": 300,
    "PET DRINK(50CL)": 350,
    "GLASS/CANNED MALT": 500,
    "FRESH YO": 600,
    "PINEAPPLE JUICE": 350,
    "MANGO JUICE": 350,
    "ZOBO DRINK": 350
}

def menu_open():
    menu_window = tk.Toplevel(root)
    menu_window.title("PAU CAFETERIA MENU")
    menu_window.geometry("350x625")
    Menu = "C:/Users/easyf/Downloads/Screenshot_23-4-2024_20274_panatlanticuniversity-my.sharepoint.com.jpeg"
    pic = Image.open(Menu)
    tk_pic = ImageTk.PhotoImage(pic)
    label = tk.Label(menu_window, image=tk_pic)
    label.Menu_Image = tk_pic
    label.pack(pady=10)
    enter_button = tk.Button(menu_window, text="PLACE ORDER", command=lambda:[order_page(), menu_window.destroy()])
    enter_button.pack()

def order_page():
    order = tk.Tk()
    order.title("PAU CAFETERIA")
    order.geometry("400x500")
    name = tk.Label(order, text="Customer Name:")
    name.pack()
    name_entry = tk.Entry(order)
    name_entry.pack()
    carb_button = tk.Button(order, text="RICE/PASTA",command=carb)
    carb_button.pack(pady=15)
    protein_button = tk.Button(order, text="PROTEINS",command=protein )
    protein_button.pack(pady=15)
    swallow_button = tk.Button(order, text="SOUPS & DISHES",command=swallow )
    swallow_button.pack(pady=15)
    side_button = tk.Button(order, text="SIDES DISHES",command=side )
    side_button.pack(pady=15)
    bev_button = tk.Button(order, text="BEVERAGES", command=beverages)
    bev_button.pack(pady=15)

def carb():
    carb_page = tk.Toplevel()
    carb_page.title("RICE/PASTA PAGE")
    carb_page.geometry("300x300")
    jollof_button = tk.Button(carb_page, text="JOLLOF RICE", command=lambda: amount("JOLLOF RICE"))
    jollof_button.pack(pady=10)
    pasta_button = tk.Button(carb_page, text="PASTA", command=lambda: amount("PASTA"))
    pasta_button.pack(pady=10)
    fried_button = tk.Button(carb_page, text="COCONUT FRIED RICE", command=lambda: amount("COCONUT FRIED RICE"))
    fried_button.pack(pady=10)
    

def protein():
    protein_page = tk.Toplevel()
    protein_page.title("PROTEIN PAGE")
    protein_page.geometry("300x300")
    sweet_button = tk.Button(protein_page, text="SWEET CHILLI CHICKEN", command=lambda: amount("SWEET CHILLI CHICKEN"))
    sweet_button.pack(pady=10)
    
    grill_button = tk.Button(protein_page, text="GRILLED CHICKEN WINGS", command=lambda: amount("GRILLED CHICKEN WINGS"))
    grill_button.pack(pady=10)
    friedbeef_button = tk.Button(protein_page, text="FRIED BEEF", command=lambda: amount("FRIED BEEF"))
    friedbeef_button.pack(pady=10)
    friedfish_button = tk.Button(protein_page, text="FRIED FISH", command=lambda: amount("FRIED FISH"))
    friedfish_button.pack(pady=10)
    boiled_button = tk.Button(protein_page, text="BOILED EGGS", command=lambda: amount("BOILED EGGS"))
    boiled_button.pack(pady=10)
    sauteed_button = tk.Button(protein_page, text="SAUTEED SAUSAGES", command=lambda: amount("SAUTEED SAUSAGES"))
    sauteed_button.pack(pady=10)

def side():
    side_page = tk.Toplevel()
    side_page.title("SIDES PAGE")
    side_page.geometry("300x300")
    savoury_button = tk.Button(side_page, text="SAVOURY BEANS", command=lambda: amount("SAVOURY BEANS"))
    savoury_button.pack(pady=10)
    roast_button = tk.Button(side_page, text="ROASTED SWEET POTATOES", command=lambda: amount("ROASTED SWEET POTATOES"))
    roast_button.pack(pady=10)
    mixed_button = tk.Button(side_page, text="MIXED VEGETABLES", command=lambda: amount("MIXED VEGETABLES"))
    mixed_button.pack(pady=10)
    boiledyam_button = tk.Button(side_page, text="BOILED YAM", command=lambda: amount("BOILED YAM"))
    boiledyam_button.pack(pady=10)
    friedplan_button = tk.Button(side_page, text="FRIED PLANTAIN", command=lambda: amount("FRIED PLANTAIN"))
    friedplan_button.pack(pady=10)

def swallow():
    swallow_page = tk.Toplevel()
    swallow_page.title("SIDES PAGE")
    swallow_page.geometry("300x300")
    eba_button = tk.Button(swallow_page, text="EBA", command=lambda: amount("EBA"))
    eba_button.pack(pady=10)
    poundo_button = tk.Button(swallow_page, text="POUNDO YAM", command=amount("POUNDO YAM"))
    poundo_button.pack(pady=10)
    atama_button = tk.Button(swallow_page, text="ATAMA SOUP", command=lambda: amount("ATAMA SOUP"))
    atama_button.pack(pady=10)
    egusi_button = tk.Button(swallow_page, text="EGUSI SOUP", command= lambda: amount("EGUSI SOUP"))
    egusi_button.pack(pady=10)

def beverages():
    bev_page = tk.Toplevel()
    bev_page.title("SIDES PAGE")
    bev_page.geometry("500x500")
    water_button = tk.Button(bev_page, text="WATER", command= lambda: amount("WATER"))
    water_button.pack(pady=10)
    glass_button = tk.Button(bev_page, text="GLASS DRINK (35CL)", command=lambda: amount("GLASS DRINK (35CL)"))
    glass_button.pack(pady=10)
    pet1_button = tk.Button(bev_page, text="PET DRINK (35CL)", command=lambda: amount("PET DRINK (35CL)"))
    pet1_button.pack(pady=10)
    pet2_button = tk.Button(bev_page, text="PET DRINK (50CL)", command=lambda: amount("PET DRINK (50CL)"))
    pet2_button.pack(pady=10)
    malt_button = tk.Button(bev_page, text="GLASS/CANNED MALT", command=lambda: amount("GLASS/CANNED MALT"))
    malt_button.pack(pady=10)
    fresh_button = tk.Button(bev_page, text="FRESH YO", command=lambda: amount("FRESH YO"))
    fresh_button.pack(pady=10)
    pin_button = tk.Button(bev_page, text="PINEAPPLE JUICE", command=lambda: amount("PINEAPPLE JUICE"))
    pin_button.pack(pady=10)
    mango_button = tk.Button(bev_page, text="MANGO JUICE", command=lambda: amount("MANGO JUICE"))
    mango_button.pack(pady=10)
    zobo_button = tk.Button(bev_page, text="ZOBO DRINK", command=lambda: amount("ZOBO DRINK"))
    zobo_button.pack(pady=10)

def amount(food_item):   
    quantity_window = tk.Toplevel()
    quantity_window.title("Quantity Page")
    quantity_window.geometry("200x200")
                
    def calculate_total():
        quantity = int(quantity_entry.get())
        price_per_item = food_prices[food_item.upper()]
        total_price = quantity * price_per_item
        
        if total_price <= 2500:
            discount_price = total_price - (10/100 * total_price)
            messagebox.showinfo("Total Price", f"Total Price for {quantity} {food_item}( 10% discount): N{discount_price}")

        elif total_price > 2500 and total_price <= 5000:
            discount_price = total_price - (15/100 * total_price)
            messagebox.showinfo("Total Price", f"Total Price for {quantity} {food_item}(15% discount): N{discount_price}")

        elif total_price < 5000 :
            discount_price = total_price - (25/100 * total_price)
            messagebox.showinfo("Total Price", f"Total Price for {quantity} {food_item}( 25% discount): N{discount_price}")

        elif total_price < 1000 :
            messagebox.showinfo("Total Price", f"Total Price for {quantity} {food_item}( no discount): N{total_price}")
        
            

    label_3 = tk.Label(quantity_window, text="HOW MANY DO YOU WANT?")
    label_3.pack()
    quantity_entry = tk.Entry(quantity_window)
    quantity_entry.pack()
    finish = tk.Button(quantity_window, text="ORDER", command=calculate_total )
    finish.pack(pady=10)

 




#Create a window for the image

root = tk.Tk()
root.title("PAU CAFETERIA")
root.geometry("315x100")
label_1 = tk.Label(root, text="WELCOME TO PAU CAFETARIA\nWHERE DELICACIES ARE SERVED FRESH AND HOT AND AT \n A GO!!!")
label_1.pack()
label_1.config(bg="gold")
root.config(bg = "gold")
go_button = tk.Button(root, text="ENTER", command=menu_open)
go_button.pack()

 
    



root.mainloop()
```

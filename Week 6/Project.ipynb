```python
#Project 2
import tkinter as tk
from tkinter import messagebox

#create main window 
root = tk.Tk()
root.title("A & S DELIVERY SERVICES")
root.geometry("500x500")
label_1 = tk.Label(root, text=f"Welcome to A & S DELIVERY SERVICE\nPlease Input your name and the weight of the product")
label_1.pack()

#add label widget for information entry
name_label = tk.Label(root, text="Name:")
name_label.pack()
name_entry = tk.Entry(root)
name_entry.pack()

weight_label = tk.Label(root, text="Weight (kg):")
weight_label.pack()
weight_entry = tk.Entry(root)
weight_entry.pack()

location_label = tk.Label(root, text="Drop Off location:")
location_label.pack()
label_entry = tk.Entry(root)
label_entry.pack()




#create a submit function
def submit():
    name = name_entry.get()
    weight = float(weight_entry.get())
    location = label_entry.get()
    
    if weight >= 10.0  and location == "Ibeju-Lekki":
        window = tk.Toplevel()
        window.title("A & S DELIVERY SERVICES" )
        window.geometry("300x300")
        label_1 = tk.Label(window, text=f"The total price of the delivery will be 4000 Naira")
        label_1.pack()
        
    if weight <= 9.9 and location == "Ibeju-Lekki":
         window = tk.Toplevel()
         window.title("A & S DELIVERY SERVICES" )
         window.geometry("300x300")
         label_1 = tk.Label(window, text=f"The total price of the delivery will be 3000 Naira")
         label_1.pack()


    if weight >= 10.0 and location == "Epe":
          window = tk.Toplevel()
          window.title("A & S DELIVERY SERVICES" )
          window.geometry("300x300")
          label_1 = tk.Label(window, text=f"The total price of the delivery will be 10,000 Naira")
          label_1.pack()
          
    if weight <= 9.9 and location == "Epe":
          window = tk.Toplevel()
          window.title("A & S DELIVERY SERVICES" )
          window.geometry("300x300")
          label_1 = tk.Label(window, text=f"The total price of the delivery will be 10,000 Naira")
          label_1.pack()

    else :
          messagebox.showerror("ERROR""ERROR IN LOCATION OR WEIGHT") 
          
        

#create the submit button
submit_button = tk.Button(root, text="Submit",command=submit)
submit_button.pack()


root.mainloop()


```
